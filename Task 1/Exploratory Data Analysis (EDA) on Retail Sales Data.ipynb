```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('menu.csv')
```

```python
#. Basic Inspection
print("Shape of the dataset:", df.shape)
print("\nColumn Names:\n", df.columns)
print("\nSummary Statistics:\n", df.describe())
```

```python
#Checking for missing values
print("\nMissing Values:\n", df.isnull().sum())
```

```python
#Distribution of Calories
plt.figure(figsize=(10,6))
sns.histplot(df['Calories'], bins=30, kde=True, color='blue')
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
#  Total Fat vs Carbohydrates
plt.figure(figsize=(10,6))
sns.scatterplot(x='Total Fat', y='Carbohydrates', hue='Category', data=df)
plt.title('Total Fat vs Carbohydrates')
plt.xlabel('Total Fat (g)')
plt.ylabel('Carbohydrates (g)')
plt.show()
```

```python
#  Calories by Category
plt.figure(figsize=(10,6))
sns.boxplot(x='Category', y='Calories', data=df)
plt.title('Calories by Food Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation Matrix
# Select only the numerical columns for correlation
numeric_cols = df.select_dtypes(include=['float64', 'int64'])

# Correlation Matrix
plt.figure(figsize=(12,8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Nutritional Values')
plt.show()
```

```python
#Average Nutritional Values by Category
# Select only the numeric columns for aggregation
numeric_cols = ['Calories', 'Total Fat', 'Carbohydrates', 'Protein']  

# Group by 'Category' and calculate the mean for the numeric columns
avg_nutrition_by_category = df.groupby('Category')[numeric_cols].mean()
print("\nAverage Nutritional Values by Category:\n", avg_nutrition_by_category)
```

```python
#Pairplot for selected nutritional columns
sns.pairplot(df[['Calories', 'Total Fat', 'Carbohydrates', 'Protein']])
plt.show()
```

```python
#Scatter plot of Calories vs Total Fat
plt.figure(figsize=(10,6))
sns.scatterplot(x='Calories', y='Total Fat', hue='Category', data=df)
plt.title('Calories vs Total Fat by Category')
plt.xlabel('Calories')
plt.ylabel('Total Fat (g)')
plt.show()
```

```python
#Scatter plot of Calories vs Protein
plt.figure(figsize=(10,6))
sns.scatterplot(x='Calories', y='Protein', hue='Category', data=df)
plt.title('Calories vs Protein by Category')
plt.xlabel('Calories')
plt.ylabel('Protein (g)')
plt.show()
```

```python
#Histograms for numerical columns
num_cols = ['Calories', 'Total Fat', 'Carbohydrates', 'Protein', 'Sugars']
df[num_cols].hist(figsize=(12,10), bins=20, color='skyblue')
plt.suptitle('Histograms for Nutritional Information', fontsize=16)
plt.show()
```
